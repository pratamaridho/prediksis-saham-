```python
import yfinance as yf

# Contoh ambil data BBCA (BEI = .JK)
ticker = "BBCA.JK"
data = yf.download(ticker, start="2000-01-01", end="2025-10-30")
```

```python
data.to_csv("BBCA.csv")
```

```python
import pandas as pd

cols = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']

df = pd.read_csv('BBCA.csv', skiprows=3, names=cols)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.info()
```

```python
df.set_index('Date', inplace=True)
```

```python
df = df.sort_index()
```
